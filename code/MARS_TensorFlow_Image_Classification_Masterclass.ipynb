# 📌 Output Size Formula

For convolution:




Convolution output formula:

## 📌 Convolution Output Formula

$$
Output = \frac{W - F + 2P}{S} + 1
$$

Where:

- \(W\) = input size
- \(F\) = filter size
- \(P\) = padding
- \(S\) = stride

Where:

- W = input size
- F = filter size
- P = padding
- S = stride

Example:

Input = 32  
Filter = 3  
Padding = 1  
Stride = 1

Output:



````

# Example convolution output calculation

W = 32
F = 3
P = 1
S = 1

output = ((W - F + 2 * P) / S) + 1
print("Output Size:", output)
````


# 🧠 SAME vs VALID Padding

## VALID Padding

No padding.

---

# Pooling

AveragePooling:

Takes average value.

# 🧱 Section 2 — Building LeNet5 from Scratch

LeNet5 is one of the earliest CNN architectures.

It teaches:
- convolution
- pooling
- flattening
- dense layers

Perfect for learning.


````

# =========================================
# LeNet5 Architecture
# =========================================

def build_lenet5(input_shape=(32, 32, 3), num_classes=10):

    model = keras.Sequential([

        layers.Conv2D(
            6,
            kernel_size=(5,5),
            activation='relu',
            input_shape=input_shape
        ),

        layers.AveragePooling2D(),

        layers.Conv2D(
            16,
            kernel_size=(5,5),
            activation='relu'
        ),

        layers.AveragePooling2D(),

        layers.Flatten(),

        layers.Dense(120, activation='relu'),
        layers.Dense(84, activation='relu'),
        layers.Dense(num_classes, activation='softmax')

    ])

    return model

model = build_lenet5()

model.summary()
````


# 🏗️ Section 3 — Designing Your Own CNN

This section is VERY important.

A real AI engineer should know:

- when to increase channels
- when to downsample
- how deep to go
- how to reduce overfitting

---

# Typical Pattern

Conv → BatchNorm → ReLU → Pool

Repeated multiple times.

---

# 📌 Why Channels Increase?

Early layers:
- detect edges

Deep layers:
- detect complex features

Therefore:
- spatial size decreases
- channels increase

Example:

32 → 64 → 128 → 256

---

# 📌 Why BatchNorm?

Benefits:

- stabilizes training
- faster convergence
- allows higher learning rate

---

# 📌 Why Dropout?

Randomly disables neurons.

Helps prevent overfitting.


````

# =========================================
# Custom CNN
# =========================================

def build_custom_cnn(
    input_shape=(224,224,3),
    num_classes=2
):

    inputs = keras.Input(shape=input_shape)

    x = layers.Conv2D(32, 3, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(64, 3, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.MaxPooling2D()(x)

    x = layers.Conv2D(128, 3, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(
        num_classes,
        activation='softmax'
    )(x)

    model = keras.Model(inputs, outputs)

    return model

custom_model = build_custom_cnn()

custom_model.summary()
````

# 🔥 Data Augmentation

Very important in real-world AI.

Why?

Because real images vary:
- lighting
- angle
- blur
- rotation
- occlusion

Augmentation improves robustness.


````

# =========================================
# DATA GENERATOR
# =========================================

IMAGE_SIZE = 224
BATCH_SIZE = 32

train_dir = "/content/dataset/train"
valid_dir = "/content/dataset/valid"
test_dir = "/content/dataset/test"

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)

valid_datagen = ImageDataGenerator(
    rescale=1./255
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

NUM_CLASSES = len(train_generator.class_indices)

print(train_generator.class_indices)
````


# 🚀 Section 5 — Transfer Learning

Transfer Learning = use pretrained knowledge.

Pretrained models already learned:

- edges
- textures
- object structure
- visual semantics

from ImageNet.

---

# 📌 Why Transfer Learning Works

Even if your task differs from ImageNet:

low-level visual features are still useful.

---

# Popular Models

| Model | Strength |
|---|---|
| ResNet | stable, strong baseline |
| DenseNet | feature reuse |
| EfficientNet | parameter efficient |
| ViT | transformer-based |


````

# =========================================
# MODEL FACTORY
# =========================================

def create_model(
    model_name='resnet',
    num_classes=2,
    input_shape=(224,224,3),
    freeze_backbone=True
):

    if model_name == 'resnet':

        backbone = ResNet50(
            include_top=False,
            weights='imagenet',
            input_shape=input_shape
        )

    elif model_name == 'densenet':

        backbone = DenseNet121(
            include_top=False,
            weights='imagenet',
            input_shape=input_shape
        )

    elif model_name == 'efficientnet':

        backbone = EfficientNetB0(
            include_top=False,
            weights='imagenet',
            input_shape=input_shape
        )

    elif model_name == 'custom':

        return build_custom_cnn(
            input_shape=input_shape,
            num_classes=num_classes
        )

    else:
        raise ValueError("Unknown model")

    backbone.trainable = not freeze_backbone

    inputs = keras.Input(shape=input_shape)

    x = backbone(inputs, training=False)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(
        num_classes,
        activation='softmax'
    )(x)

    model = keras.Model(inputs, outputs)

    return model
````


# 🧊 Freeze vs Fine-Tune

## Freeze Backbone

Only train classifier head.

Benefits:
- fast
- stable
- good for small datasets

---

## Fine-Tuning

Unfreeze some layers.

Benefits:
- adapts features to your dataset
- often higher accuracy

Risks:
- overfitting
- catastrophic forgetting

---

# 🔥 Real-World Tip

Stage training:

1. Freeze backbone
2. Train head
3. Unfreeze last blocks
4. Fine-tune with lower LR



# 🎯 Section 6 — Loss Functions

Loss function = optimization target.

---

# Classification Losses

## Categorical Crossentropy

Most common.

## Binary Crossentropy

Binary classification.

## Focal Loss

Great for imbalance.

Focuses on hard samples.

---

# 📌 When to Use Focal Loss

Example:

95% normal images  
5% damaged images

Without focal loss:
model may ignore minority class.


````

# =========================================
# CUSTOM FOCAL LOSS
# =========================================

def focal_loss(gamma=2., alpha=0.25):

    def loss(y_true, y_pred):

        epsilon = 1e-7

        y_pred = tf.clip_by_value(
            y_pred,
            epsilon,
            1. - epsilon
        )

        cross_entropy = -y_true * tf.math.log(y_pred)

        weight = alpha * tf.pow(1 - y_pred, gamma)

        loss_value = weight * cross_entropy

        return tf.reduce_mean(tf.reduce_sum(loss_value, axis=1))

    return loss

# Example
loss_fn = focal_loss()
````


# ⚖️ Handling Imbalanced Dataset

# Techniques

## 1. Class Weight

Increase penalty for minority class.

## 2. Focal Loss

Focus hard samples.

## 3. Oversampling

Duplicate minority samples.

## 4. Better Data

Usually the BEST solution.

---

# 🔥 Important

Data quality > fancy architecture.


````

# =========================================
# TRAIN CONFIG
# =========================================

MODEL_NAME = 'efficientnet'
FREEZE_BACKBONE = True

USE_FOCAL_LOSS = True

EPOCHS = 5
LEARNING_RATE = 1e-3

# =========================================
# CREATE MODEL
# =========================================

model = create_model(
    model_name=MODEL_NAME,
    num_classes=NUM_CLASSES,
    freeze_backbone=FREEZE_BACKBONE
)

# =========================================
# LOSS
# =========================================

if USE_FOCAL_LOSS:
    loss_function = focal_loss()
else:
    loss_function = 'categorical_crossentropy'

# =========================================
# COMPILE
# =========================================

model.compile(
    optimizer=keras.optimizers.Adam(
        learning_rate=LEARNING_RATE
    ),
    loss=loss_function,
    metrics=['accuracy']
)

model.summary()
````

````

# =========================================
# CALLBACKS
# =========================================

callbacks = [

    EarlyStopping(
        patience=5,
        restore_best_weights=True
    ),

    ReduceLROnPlateau(
        factor=0.2,
        patience=2
    ),

    ModelCheckpoint(
        'best_model.keras',
        save_best_only=True
    )

]
````

````

# =========================================
# TRAIN
# =========================================

history = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=EPOCHS,
    callbacks=callbacks
)
````

````

# =========================================
# FINE-TUNING EXAMPLE
# =========================================

# Unfreeze last layers

for layer in model.layers[-20:]:
    layer.trainable = True

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss=loss_function,
    metrics=['accuracy']
)

history_finetune = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=3
)
````


# 📊 Section 7 — Evaluation

Accuracy alone is NOT enough.

Need:

- Precision
- Recall
- F1-score
- Confusion Matrix

Especially for imbalanced data.


````

# =========================================
# EVALUATION
# =========================================

test_datagen = ImageDataGenerator(
    rescale=1./255
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=1,
    shuffle=False,
    class_mode='categorical'
)

pred_probs = model.predict(test_generator)

y_pred = np.argmax(pred_probs, axis=1)
y_true = test_generator.classes

acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print("Accuracy :", acc)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
````

````

# =========================================
# CONFUSION MATRIX
# =========================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,8))
plt.imshow(cm)

plt.title("Confusion Matrix")
plt.colorbar()

plt.xlabel("Predicted")
plt.ylabel("True")

plt.show()
````

````

# =========================================
# SHOW WRONG PREDICTIONS
# =========================================

class_names = list(test_generator.class_indices.keys())

wrong_indices = np.where(y_true != y_pred)[0]

plt.figure(figsize=(15,15))

for i, idx in enumerate(wrong_indices[:9]):

    img_path = test_generator.filepaths[idx]

    img = keras.utils.load_img(
        img_path,
        target_size=(IMAGE_SIZE, IMAGE_SIZE)
    )

    plt.subplot(3,3,i+1)

    plt.imshow(img)

    plt.title(
        f"True: {class_names[y_true[idx]]}\n"
        f"Pred: {class_names[y_pred[idx]]}"
    )

    plt.axis('off')

plt.tight_layout()
plt.show()
````
